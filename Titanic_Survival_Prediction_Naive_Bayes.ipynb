```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic_data.csv')
df.head()
```

```python
df = df[['Survived','Pclass','Sex','Age','Fare']]
```

```python
df.head()
```

```python
df.shape
```

```python
df['Sex'] = df['Sex'].map({'female':0,'male':1}).astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.drop('Survived',axis=1)
```

```python
X.head()
```

```python
Y = df['Survived']
Y.head()
```

```python
X.Age = X.Age.fillna(X.Age.mean())
X.head()
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)
```

```python
Pclass = int(input('Enter person class'))
gender = int(input('Enter person gender 0-female or 1-male'))
age = int(input('Enter the age of person'))
fare = float(input('Enter persons fare'))
person = [[Pclass,gender,age,fare]]
result = model.predict(person)
print(result)

if result == 1:
  print('the person might be Survived')
else:
  print('the person might not be survived')
```

```python
Y_pred = model.predict(X_test)
print(np.column_stack((Y_pred,Y_test)))
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy of the model : {0}%'.format(accuracy_score(Y_test,Y_pred)*100))
```
